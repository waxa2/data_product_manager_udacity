```python
import pandas as pd
```

```python
df_ride = pd.read_csv('taxi_rides.csv')
df_ride.head()
```

```python
df_zip = pd.read_csv('zip_lat_long.csv')
df_zip.tail()# else zip 99929
```

```python

df_zip['LAT_LEAD'] = df_zip['LAT'].shift(-1)
df_zip['LNG_LEAD'] = df_zip['LNG'].shift(-1)
```

```python
df_zip.head()
```

```python
import pandas as pd

# Assuming you have two DataFrames, df_ride, and df_zip

# Create a new column 'Nueva_Columna_ZIP' in df_ride
df_ride['Nueva_Columna_ZIP'] = None

# Loop through each row in df_ride
for i in range(df_ride.shape[0]):
    # Find rows in df_zip that match the conditions
    condition = (
        (df_ride['pickup_latitude'][i] >= df_zip['LAT']) &
        (df_ride['pickup_latitude'][i] < df_zip['LAT_LEAD']) &
        (df_ride['pickup_longitude'][i] >= df_zip['LNG']) &
        (df_ride['pickup_longitude'][i] < df_zip['LNG_LEAD'])
    )
    
    # If any match is found, assign the corresponding ZIP value to the new column
    if condition.any():
        matching_index = df_zip.index[condition].tolist()[0]  # Assuming one match
        df_ride.at[i, 'Nueva_Columna_ZIP'] = df_zip.at[matching_index, 'ZIP']

# Now, df_ride will have a new column 'Nueva_Columna_ZIP' with the assigned ZIP values
```

```python
# We have two DataFrames, df_ride, and df_zip

# Create a new column 'Nueva_Columna_ZIP' in df_ride

df_ride['Nueva_Columna_ZIP_dropoff'] = None

# Loop through each row in df_ride
for i in range(df_ride.shape[0]):
    # Find rows in df_zip that match the conditions
    condition = (
        (df_ride['dropoff_latitude'][i] >= df_zip['LAT']) &
        (df_ride['dropoff_latitude'][i] < df_zip['LAT_LEAD']) &
        (df_ride['dropoff_longitude'][i] >= df_zip['LNG']) &
        (df_ride['dropoff_longitude'][i] < df_zip['LNG_LEAD'])
    )
    
    # If any match is found, assign the corresponding ZIP value to the new column
    if condition.any():
        matching_index = df_zip.index[condition].tolist()[0]  # Assuming one match
        df_ride.at[i, 'Nueva_Columna_ZIP_dropoff'] = df_zip.at[matching_index, 'ZIP']


# Now, df_ride will have a new column 'Nueva_Columna_ZIP' with the assigned ZIP values
```
